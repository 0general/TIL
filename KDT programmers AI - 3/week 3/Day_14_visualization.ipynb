#### x, y축에 설명 추가하기

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.plot(x,x**2)
plt.show()
```

#### x, y축의 범위를 설정하기

그래프의 일부만 보고싶을 때 사용할 수 있다.  
`plt.axis([x_min, x_max, y_min, y_max])`

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]

plt.plot(x,x**2)
plt.show()
```

#### x, y 축에 눈금 설정하기
  
`xticks`, `yticks` 함수를 이용해보자.  
  
tick이 눈금 하나하나를 의미한다.  
  
인자로 데이터의 컨테이너를 주면 된다. 여기서는 리스트를 넣어보자.

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 5, 1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)]) # y축의 눈금 설정

plt.plot(x,x**2)
plt.show()
```

#### 그래프에 title 달기

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 5, 1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)]) # y축의 눈금 설정

plt.title("y=x^2 graph")

plt.plot(x,x**2)
plt.show()
```

#### 그래프에 라벨 달기
  
plt.plot(x, x**2, label="trend")

### 범례 확인
라벨을 달고 나서  
plt.legend() 를 사용하면 범례가 생긴다.

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5,5,0,25]) # [x_min, x_max, y_min, y_max]
plt.xticks([i for i in range(-5, 5, 1)]) # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)]) # y축의 눈금 설정
plt.title("y=x^2 graph")

plt.plot(x,x**2, label = "trend")
plt.legend()

plt.show()
```

### 꺾은 선 그래프 (Plot)
- `.plot()`
- 가장 간단한 그래프

```python
x = np.arange(20) # 0~19
y = np.random.randint(0,20,20) # 난수를 20번 생성

x, y
```

```python
plt.plot(x,y) # 왼쪽이 x축에 해당, 오른쪽이 y축에 해당
plt.show()
```

y값이 난수였기 때문에 딱히 어떤 추세를 발견할 수 없다.

Extra : y축을 20까지 보이게 하고싶다면?, y축을 "5"단위로 보이게 하고 싶다면?
.axis(), .yticks()

```python
plt.plot(x,y)
plt.axis([0,20,0,20])
plt.yticks([0,5,10,15,20])
plt.show()
```

### 산점도 (Scatter Plot)
  * `.scatter()`
  
    
두 변수 사이의 상관관계를 확인하기 쉽다

```python
plt.scatter(x,y)
plt.show()
```

### 위의 두 그래프가 각각 어느 경우에 사용될까?

1.  꺾은선 그래프  
   ex) 시계열 - x축 : 시간, y축 : 그에 대한 변수  
   종속적이 두 개의 변수 사이의 변화를 확인하기 위해 주고 사용한다.  
  
  

2. 산점도  
   ex) 관계가 없는 두 변수  
   완전의 별개의 변수 사이에 사용한다. 관계를 모르는 두 변수에 산점도를 사용했을 때 어떤 상관관계를 발견할 수도 있다.

### 박스 그림 (Box Plot)
* 수치형 데이터에 대한 정보(Q1, Q2, Q3, min, max)  

단일 변수가 있을 때 이 변수의 전반적인 분포를 보고 싶을 때 사용한다.  
Quantile(사분위 수) 중 Q1(25%), Q2(50%), Q3(75%), min, max data를 담고 있다.

```python
plt.boxplot(y)

plt.show()
```

순서대로 min(가장 아래 선), Q1(박스의 아래), median(붉은 선), Q3(박스 위), max(가장 위 선)

```python
# Extra : Plot의 title을 "Box plot of y"
plt.boxplot(y)
plt.title("Box plot of y")

plt.show()
```

Box plot은 단일 변수말고 2개 이상의 변수를 같이 써서 비교하는데 이용하기도 한다.  
인자로 컨테이너를 넣어주자. 

```python
plt.boxplot((x,y))

plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프 (Bar Plot)

### cf) Histogram
- 도수분포를 직사각형의 막대 형태로 나타냈다.

### 커널밀도그림 (Kernel Density Plot)
- 히스토그램과 같은 연속적인 분포르 곡선화해서 그린 그림
- `sns.kdeplot()`

```python
# in Histogram

x = np.arange(0,22,2)
y = np.random.randint(0,20,20)

plt.hist(y, bins = x)
plt.show()
```

- `random.randint` : a <= x <= b
- `np.random.randint` : a <= x < b

```python
# kdeplot

sns.kdeplot(y)

plt.show() 
```

seaborn은 plt의 함수를 공유하므로 plt.show를 써줘야한다.

```python
sns.kdeplot(y, shade=True)

plt.show() 
```

### 카운트그림 (Count Plot)
- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- `sns.countplot()`

```python
vote_df = pd.DataFrame({"name": ['Andy', 'Bob', 'Cat'], "vote":[True, True, False]})

vote_df
```

#### ⬇ In matplotlib barplot


```python

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x=[False,True], height=vote_count['name'])

plt.show()
```

#### ⬇ sns의 countplot

```python
sns.countplot(x=vote_df['vote'])

plt.show()
```

sns를 사용하면 조금 더 편리하게 counting한 경과를 시각화해서 보여줄 수 있다.

### 캣그림 (Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- `sns.catplot()`, x,y,data 가 필수 3가지 인자

```python
covid = pd.read_csv("./country_wise_latest.csv")

covid
```

```python
sns.catplot(x="WHO Region", y="Confirmed", data=covid) # hue라는 인자를 넣어줄 수 있다. 이것은 범주나 수치형 데이터간의 관계를 나타낸다.
plt.show()
```

x축의 크기가 좁아서 이름이 겹치는 경우에는 sns로 만든 catplot 객체를 s로 받아온 다음,  
sns 의 figsize 조절하는 과정을 아래와 같이 거치면 된다.

```python
s = sns.catplot(x="WHO Region", y="Confirmed", data=covid)
s.fig.set_size_inches(10,6)
plt.show()
```

## Cat Plot은 어떤 그래프의 형태가 아닌 여러 개의 데이터를 묶는 그래프의 방법이다.
따라서, 위 그래프는 Catplot의 형태나 모양이 아니다.  
위 그래프는 `kind`의 default 값인 strip의 모양이고,  
모양을 바꾸려면 `kind` 파라미터를 바꿔야한다.

```python
s = sns.catplot(x="WHO Region", y="Confirmed", data=covid, kind='violin') # kind default = strip
s.fig.set_size_inches(10,6)
plt.show()
```

### 스트립그림 (Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- `sns.stripplot()`, x, y, data가 필수 3가지 인자

산점도와 마찬가지로 데이터의 수치를 점으로 표현해준다.

```python
s = sns.stripplot(x="WHO Region", y = "Recovered", data=covid)

plt.show()
```

### cf) swarmplot
- 기본적으로 strip plot과 거의 유사하다
- 하지만 strip plot처럼 동일한 value를 가진 값들이 많은 경우 옆으로 분산하여 해당하는 값들이 얼마나 있는지 확인하기 쉽게 해준다. 

```python
s = sns.swarmplot(x="WHO Region", y = "Recovered", data=covid)

plt.show()
```

오류가 발생하는 이유 : 점들의 값이 너무 크다 보니 다 그릴 수 없다는 의미로 오류 발생

### 히트맵 (Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프
- `sns.heatmap()`

이 히트맵을 가장 많이 사용하는 예시가 상관계수(Correlation 행렬)이다.

```python
# 히트맵 예제

covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```

히트맵이라는 행과 열에 대한 정보를 색깔로 일목요연하게 확인가능
